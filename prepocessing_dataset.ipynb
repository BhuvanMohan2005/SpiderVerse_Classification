Thumbs.db → Windows system file storing thumbnail cache of images in a folder.

**Output Directory:**  
  - `output_dir = "clean_dataset"` is where **preprocessed images** will be saved.  
  - `os.makedirs(output_dir, exist_ok=True)` ensures the folder exists or creates it.  

- **Image Size:**  
  - `IMG_SIZE = 224` sets a **uniform size** for all images (224 × 224 pixels) for machine learning models.  

```python
import os
import cv2
from PIL import Image

output_dir = "clean_dataset"

IMG_SIZE = 224

os.makedirs(output_dir, exist_ok=True)
```

### Dataset Preprocessing Steps

- **Input Dataset:** `spiderverse_dataset` contains subfolders for each class/label.   
- **Corrupted / Hidden Files**:.DS_Store files were detected and removed automatically.
- **Folder Check:** Ensure only directories (classes) are processed.  
- **Create Output Folder:** Preserve class structure in the cleaned dataset directory.  
- **Open and Convert Images:** Convert all images to **RGB** format for consistency.  
- **Resize Images:** Resize all images to a fixed size (`IMG_SIZE × IMG_SIZE`) suitable for ML models.  
- **Save Images:** Store the cleaned images in the output folder.  
- **Handle Corrupted Images:** Skip and remove images that cannot be opened or processed.  
- **Count Images:** Keep track of successfully processed images per class.  
- **Final Output:** Cleaned, uniform, and ready-to-use dataset with preserved class structure.

```python

input_dir = "spiderverse_dataset"

for class_name in os.listdir(input_dir):


    class_path = os.path.join(input_dir, class_name)
    if not os.path.isdir(class_path):
        continue

    save_class = os.path.join(output_dir, class_name)
    os.makedirs(save_class, exist_ok=True)

    count = 0
    for img_name in os.listdir(class_path):
        img_path = os.path.join(class_path, img_name)
        try:
            img = Image.open(img_path).convert("RGB")
            img = img.resize((IMG_SIZE, IMG_SIZE))
            save_path = os.path.join(save_class, img_name)
            img.save(save_path)
            count += 1
        except Exception as e:
            print(f"Corrupted image removed: {img_path} ({e})")

    print(f"Class '{class_name}': {count} images cleaned and saved.")
```

.DS_Store files were detected and removed automatically.

No other corrupted images were found.

Image Preprocessing:

All images were resized to 224×224 pixels.

Converted all images to RGB format for CNN compatibility.
